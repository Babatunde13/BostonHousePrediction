```python
#Selecting our features and target variable
feature = boston.drop('MEDV', axis=1)
target = boston[['MEDV']]
```

# Train Test Split
I'll be importing the train_test_split methos from the model selection module in sklearn

```python
#Importing the function, and setting the test size to 20%
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(feature, target, test_size=.2, random_state=5)
```

# Model Building And Evaluation

```python
#Importing Libraries
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
#Creating a Linear Regression class and normalizing the features
lin_reg = LinearRegression(normalize=True)

#Fitting the model
lin_reg.fit(X_train, Y_train)
```

```python
pred = lin_reg.predict(X_test)
```

```python
print(
    f' The coefficient of the RM feature is {lin_reg.coef_[0][0]} while the coefficient of the feature LSTAT is {lin_reg.coef_[0][0]} '
)

print('The intercept is ', lin_reg.intercept_[0])
```

```python
mae = mean_absolute_error(pred, Y_test)
rmse = np.sqrt(mean_squared_error(pred, Y_test))
r2 = r2_score(Y_test, pred)
print(f'The RMSE of the model is {rmse.round(3)}')
print(f"The Mean Absolute Error for the model is {mae.round(3)}")
print(f'The R2 score for the model is {r2.round(3)}')
```

```python
lin_reg.fit(X_train[['RM', 'LSTAT']], Y_train)
pred_1 = lin_reg.predict((X_test[['RM', 'LSTAT']]))
print(lin_reg.score(X_test[['RM', 'LSTAT']], Y_test))
print(mean_absolute_error(Y_test, pred_1))
mean_squared_error(Y_test, pred_1)
```


Let's use a polynomial regression to see if we can have a better model

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
import numpy as np

scale = StandardScaler()
for i in range(2, 8):
    pipe = Pipeline([
        ('scale', scale),
        ('poly', PolynomialFeatures(degree=i)),
        ('model', lin_reg)
    ])

    pipe.fit(X_train, np.array(Y_train).reshape(-1, 1))
    print(pipe.score(X_test, np.array(Y_test).reshape(-1, 1)))
    print(mean_squared_error(Y_test, pipe.predict(X_test)))
    print(mean_absolute_error(Y_test, pipe.predict(X_test)))
    print('')
```

 from the model evaluation above we noticed that degree of n = 2 happens to be the best degree to use, if we're to use Polynomial Regression.
