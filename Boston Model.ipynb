# Feature Selection and Model Selection
From the info we can see that the data does n ot have any null value and they're all float, so with this a regression  model will be a good model for this data.

```python
#Selecting our features and target variable
feature = boston[[ 'RM', 'LSTAT']]
target = boston[['MEDV']]
```

# Train Test Split
I'll be importing the train_test_split methos from the model selection module in sklearn

```python
#Importing the function, and setting the test size to 20%
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(feature, target, test_size=.2, random_state=5)
```

# Model Building And Evaluation

```python
#Importing Libraries
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
#Creating a Linear Regression class and normalizing the features
lin_reg = LinearRegression(normalize=True)

#Fitting the model
lin_reg.fit(X_train, Y_train)
```

```python
pred = lin_reg.predict(X_test)
```

```python
print(
    f' The coefficient of the RM feature is {lin_reg.coef_[0][0]} while the coefficient of the feature LSTAT is {lin_reg.coef_[0][0]} '
)

print('The intercept is ', lin_reg.intercept_[0])
```

```python
mae = mean_absolute_error(pred, Y_test)
rmse = np.sqrt(mean_squared_error(pred, Y_test))
r2 = r2_score(Y_test, pred)
print(f'The RMSE of the model is {rmse.round(3)}')
print(f"The Mean Absolute Error for the model is {mae.round(3)}")
print(f'The R2 score for the model is {r2.round(3)}')
```

Let's use a polynomial regression to see if we can have a better model

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
def poly_reg(degree):
    poly = PolynomialFeatures(degree=degree)
    x_poly = poly.fit_transform(X_train)
    lr = LinearRegression(normalize=True)
    lr.fit(x_poly, Y_train)
    y_pred = lr.predict(poly.fit_transform(X_test))
    mae = mean_absolute_error(y_pred, Y_test)
    rmse = np.sqrt(mean_squared_error(y_pred, Y_test))
    r2 = r2_score(Y_test, y_pred)
    print('================================')
    print(f'With n = {degree}')
    print(f'The RMSE of the model is {rmse.round(3)}')
    print(f"The Mean Absolute Error for the model is {mae.round(3)}")
    print(f'The R2 score for the model is {r2.round(3)}')
    print('================================')
```

```python
for i  in range(10):
    poly_reg(i)
```

Looking at the values for rmse, mae and r2 we can see that the model is best when degree is 2 at this point the model has the highest r2 score.
